Let us find out which drink has the higest number of calories

```python
fig = px.pie(df.sort_values('Calories', ascending = False).head(17), 
             values='Calories', 
             names='Beverage', 
             title='Top 7 highest Calorie drinks',
             color_discrete_sequence=px.colors.sequential.Viridis)
fig.show()
```

```python
#Top 7 calorie drinks excluding White chocolate mocha


df = df.drop(df[df['Beverage'] == 'White Chocolate Mocha (Without Whipped Cream)'].index)
fig = px.pie(df.sort_values('Calories', ascending = False).head(17), 
             values='Calories', 
             names='Beverage', 
             title='Top 7 highest Calorie drinks',
             color_discrete_sequence=px.colors.sequential.Viridis)
fig.show()
```

How much caffine and carbohydrates are you getting 

```python
fig = px.sunburst(df.sort_values('Calories', ascending = False).head(10), 
             values='Calories', path = ['Caffeine (mg)', ' Total Carbohydrates (g) '],
             title='Sunburst Plot',
             color_discrete_sequence=px.colors.sequential.Viridis)
fig.show()
```
